# 2. Background Theory

## 2.1 Theorem: Singular Value Theorem for Linear Transformations 
Source: Friedberg, Insel & Spence, Linear Algebra, Theorem 6.26, p. 403.

Let $F^n$ and $F^m$ be finite-dimensional inner product spaces. Then there exist orthonormal bases  $\beta = \{v_1, v_2, \ldots, v_n\} $ for $ F^n $ and $\gamma = \{u_1, u_2, \ldots, u_m\} $ for  $F^m$ with positive scalars $\sigma_1 \geq \sigma_2 \geq \cdots \geq \sigma_r$ such that $T(v_i) = \sigma_i u_i$  if$ 1 \leq i \leq r $ and $ T(v_i) = 0 $ if $ i > r $.


## 2.2 Definition of Singular Value Decomposition (SVD)
Source: Adapted from Friedberg, Insel & Spence, Linear Algebra.

Let $A$ be an $m \times n$ matrix of rank $r$,representing a linear transformation from $ \mathbb{R}^n $ to $ \mathbb{R}^m $. The Singular Value Decomposition (SVD) breaks down $ A$ into the form: 
$
A = U \Sigma V^T
$

Where:

- $U$ is an $m \times m$ matrix with orthonormal columns
- $\Sigma$ is an $m \times n$ diagonal matrix containing singular values $\sigma_1 \ge \cdots \ge \sigma_r > 0$ on the main diagonal
- $V$ is an $n \times n$ matrix with orthonormal columns

The SVD reveals the geometric action of $A$: any linear transformation can be decomposed into a rotation or reflection $ (V^T)$, scalings along orthogonal directions $(\Sigma)$, and another rotation/reflection $(U)$.

## 2.3 Proof: Singular Value Decomposition Theorem for Matrices 
Source: Friedberg, Insel & Spence, Linear Algebra, Theorem 6.27, p. 406.

Let $A$ be an $ m \times n $ matrix of rank $r $ with the positive singular values $ \sigma_1 \geq \sigma_2 \geq \cdots \geq \sigma_r $, and let $ \Sigma $ be the $ m \times n $ matrix defined by
$
\Sigma_{ij} = 
\begin{cases} 
\sigma_i & \text{if } i = j \leq r \\
0 & \text{otherwise}. 
\end{cases}
$
Then there exists an $ m \times m $ unitary matrix $ U $ and an $n \times n $ unitary matrix $ V $ such that
$
A = U\Sigma V^T
$

${Proof}.$
Let $ T = L_A : F^n \rightarrow F^m $. By Theorem 6.26, there exist orthonormal bases $ \beta = \{v_1, v_2, \ldots, v_n\} $ for $ F^n $ and $ \gamma = \{u_1, u_2, \ldots, u_m\} $ for $ F^m $ such that $ T(v_i) = \sigma_i u_i $ for $ 1 \leq i \leq r $ and $ T(v_i) = 0 $ for $ i > r $. Let $ U$ be the $ m \times m $ matrix whose $j$th column is $ u_j$ for all $j$, and let $V$ be the $ n \times n $ matrix whose $j$th column is $v_j$for all $j$. 

The $j$th column of $AV$ is $Av_j = \sigma_j u_j$. The $j$th column of $\Sigma$ is $\sigma_j e_j$, where $e_j$ is the $j$th standard vector of $F^m$. So by Theorem 2.13(a), the $j$th column of $U\Sigma $ is $U(\sigma_j e_j)$, by the property of linear transformation of matrix, we can get $
U(\sigma_j e_j)= \sigma_j U e_j$ and by Theorem 2.13(b ) $ \sigma_j U e_j= \sigma_j u_j$. Thus, $U(\sigma_j e_j)= \sigma_j u_j$.

Since $ AV $ and $ U\Sigma $ are $ m \times n $ matrices, hence $AV = U\Sigma $. Therefore $ A = AVV^T = U\Sigma V^T$.$\square$

# 4. Applying SVD to Image Compression

Singular Value Decomposition is widely applied to image storage in real-world applications. A digital image can be seen as a matrix where each entry is the value of a pixel. For an $m \times n$ grayscale image, an $m \times n$ matrix $A$ can be stored and processed.

In the background theory, any matrix $A$ (representing an image) can be decomposed as $A = U \Sigma V^T$ by using Singular Value Decomposition. Following Abdi (2006), the matrix $ A $ reconstructed from the first $K$ eigenvectors is called $A_K$:
$
A_K = U_K \Sigma_K V_K^T = \sum_{i=1}^{K} \sigma_i \mathbf{u}_i \mathbf{v}_i^T
$
$\Sigma_K$ contains the first $K$ singular values. When $K$ is small, which means we keep fewer singular values and the image is compressed with low quality. This approach reduces the storage requirement for the image effectively.
As explained in Abdi (2006), the rank of matrix $A$ is the number of non-zero singular values in $\Sigma$, which is $K$. This property is fundamental to image compression. 
The null space of matrix $A$ is the span of the right singular vectors in $V$ corresponding to zero singular values. It represents the patterns that are absent from the image.

Then implement the Singular Value Decomposition for image compression using Python code in a Jupyter Notebook. The implementation allows users to upload an image from their device and compress it by changing the value of $K$
(the number of singular values). By comparing the compressed images with the original through different 
$K$ values. Observing that smaller $K$ values result in poorer image quality but require less storage space.

First, let's try to use Python Code to do Singular Value Decomposition on the matrix which is provided above as an example, the matrix is $A = \begin{bmatrix} 4 & 2 & 1 \\ 2 & 4& 1 \end{bmatrix}$

```python
# Code for matrix computation adapted from Luo (2019).
import numpy as np
from scipy.linalg import svd


# decompose SVD on matrix by using Python code
X = np.array([[4, 2, 1], [2, 4, 1]])
print(X)

U, singular, V_transpose = svd(X)
print("U:", U)
print("Singular array",singular)
print("V^{T}",V_transpose)
```

After trying SVD on a single matrix, we now apply SVD to compress a real image. Following the code below, upload your own image from your device and retrieve your pixel data for the image, which is represented as a matrix. Then, convert the RGB image to a grayscale image and apply SVD to the resulting matrix. By changing the value of K to compress the image. K shows the number of singular values in the compressed image. Comparing compressed images with the original image, it is evident that when K=5, the image has the lowest quality, indicating that its storage is very small. And when K becomes larger, the image will become clearer, and the storage space for the image will increase. 

```python
# Code adapted from Dey (2018), Hands-On Image Processing with Python.

# upload image from the device
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from scipy.linalg import svd

im = mpimg.imread('luo.jpg')

print(im.shape, im.dtype, type(im))

plt.figure(figsize=(10,10))
plt.imshow(im)
plt.axis('off')
plt.show()

#convert RGB image into grayscale
def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray_cat = 0.2989 * r + 0.5870 * g + 0.1140 * b

    return gray_cat
gray_cat = rgb2gray(im) 

# Image compression code adapted from Selvam & Selvam (2021).
U, S, V_T = svd(gray_cat, full_matrices=False)
S_diag = np.diag(S)


# show different k values  
def svd_compress(U, S, V_T, k):
    return U[:, :k] @ np.diag(S[:k]) @ V_T[:k, :]


fig, axes = plt.subplots(5, 2, figsize=(10, 20))
ks = [5, 10, 20, 100, 200]

for i, k in enumerate(ks):
    
    compressed_image = svd_compress(U, S, V_T, k)

    axes[i, 0].imshow(compressed_image, cmap='gray')
    axes[i, 0].set_title(f"k = {k}")
    axes[i, 0].axis('off')

    axes[i, 1].imshow(gray_cat, cmap='gray')
    axes[i, 1].set_title("Original Image")
    axes[i, 1].axis('off')

plt.tight_layout()
plt.show()

```
